```python
import pickle
filehandler = open('forcing_47_minus20.obj', 'rb') 
forx = pickle.load(filehandler)
```

```python
import numpy as np
import xsimlab as xs
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
ASTroCATclone
```

```python
# RUN 28

model_in = xs.create_setup(
        model=ASTroCATclone,
    clocks={
         'time': np.linspace(1,365*15,365*15*60)  # 10*365,10*365*9)
     },
    input_vars={
        'N__initVal':1,    # inital Nutrient amount
        'N__supplyrate':5, # nutrient supply rate
        
        'P__initVal': 1,   # inital total biomass of Phytoplankton
        'P__PMinEsd': .5,   # minimal Phytoplankton Size Spectral range
        'P__PMaxEsd': 50,  # minimal Phytoplankton Size Spectral range
        'P__NP': 20,       # number of Phytoplankton Types (=num of Zooplankton Types)
        'P__m': .05,        # mortality rate of Phytoplankton
        'P__kw': 0.4,
        'P__kc': 0.3,
        'P__OptI':10,
        
        'Z__initVal': 1,   # inital total biomass of Zooplankton
        'Z__zeta': .1,      # mortality rate quadratic
        'Z__deltaxprey': .25,  # log10 prey size tolerance
        'Z__KsZ': 3,       # grazing half saturation constant  # 3 '"'"'"'"'"'"':!:!:!:!:!::!:!:!::!:!:!::!!'"'"'"'"'"'"'
        'Z__f_eg': .33,    # egested food
        'Z__epsilon': .33,  # assimilated food
        
        'D__initVal': 0.1,  # initial Detritus Biomass
        'D__remin_rate': 0.1,  # Detritus remineralization rate
        
        'env__kappa':0.1  # diffusive mixing constant
    },
    output_vars={
        'env__component': None,
        'P__size': None,
        'Z__size': None,
        'env__stateout': 'time'
    }
)



model_out = model_in.xsimlab.run(model=ASTroCATclone)


#model_out.to_netcdf('output_05.nc')


new_plot(model_out,forx, 20)

plt.savefig('ONR1_output_28.pdf', bbox_inches='tight')
```

```python
NP = 20 # number of phytoplankton
PMinEsd = 0.5
PMaxEsd = 50

numbers = np.array([i for i in range(NP)])
#sizes = (np.log(self.PMaxEsd) - np.log(self.PMinEsd))* numbers / (self.NP-1) + np.log(self.PMinEsd)
sizes = (PMaxEsd - PMinEsd)* numbers / (NP-1) + PMinEsd

logsizes = np.exp((np.log(PMaxEsd) - np.log(PMinEsd))* numbers / (NP-1) + np.log(PMinEsd))
logsizes
```

```python
SIZE = logsizes
```

```python
P_pico = SIZE[SIZE<=2]
mask = ((SIZE > 2) & (SIZE <= 20))
P_nano = SIZE[mask]
mask2 = ((SIZE > 20) & (SIZE <= 50))
P_micro = SIZE[mask2]
```
