```python
import numpy
import matplotlib.pyplot as plt
from scipy.integrate import cumtrapz

mean = 0
std = 1 
num_samples = 1000
samples = numpy.random.normal(mean, std, size=num_samples)

plt.plot(samples)
plt.show()
```

```python
print('Mean: ',numpy.mean(samples))
print('Median: ',numpy.mean(samples))
print('Var: ',numpy.var(samples))
```

```python
int1 = cumtrapz(samples)
plt.plot(int1)
plt.show()
```

```python
int2 = cumtrapz(int1)
plt.plot(int2)
plt.show()
```

```python
int3 = cumtrapz(int2)
plt.plot(int3)
plt.show()
```
